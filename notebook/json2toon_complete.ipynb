## 1. Introduction <a id='introduction'></a>

JSON2TOON is a format converter that transforms JSON into a more compact, human-readable TOON format.

### Key Features:
- **Compact**: Reduces token usage by up to 40%
- **Human-readable**: Easy to read and write
- **Lossless**: Perfect round-trip conversion
- **Configurable**: Multiple formatting options
- **Robust**: Handles all edge cases

## 2. Setup and Installation <a id='setup'></a>

```python
# Import required modules
import sys
import json
from pathlib import Path

# Add parent directory to path for imports
sys.path.insert(0, str(Path.cwd().parent / 'src'))

from json2toon import ToonEncoder, ToonDecoder, ToonConfig
from json2toon.core import json_to_toon, toon_to_json, get_conversion_stats

print("âœ“ All modules imported successfully")
```

## 3. Core Encoding <a id='encoding'></a>

The encoder converts JSON data to TOON format.

```python
# Create encoder instance
encoder = ToonEncoder()

# Simple dictionary
data = {"name": "Alice", "age": 30, "active": True}

toon = encoder.encode(data)
print("JSON:")
print(json.dumps(data, indent=2))
print("\nTOON:")
print(toon)
```

```python
# Array encoding
data = {"numbers": [1, 2, 3, 4, 5]}

toon = encoder.encode(data)
print("Array Example:")
print(toon)
```

```python
# Nested object encoding
data = {
    "user": {
        "name": "Bob",
        "address": {
            "city": "NYC",
            "zip": "10001"
        }
    }
}

toon = encoder.encode(data)
print("Nested Object Example:")
print(toon)
```

## 4. Core Decoding <a id='decoding'></a>

The decoder converts TOON format back to JSON data.

```python
# Create decoder instance
decoder = ToonDecoder()

# Decode simple TOON
toon_str = """name: Alice
age: 30
active: true"""

data = decoder.decode(toon_str)
print("Decoded Data:")
print(json.dumps(data, indent=2))
```

```python
# Decode array
toon_str = """numbers:
  [1, 2, 3, 4, 5]"""

data = decoder.decode(toon_str)
print("Decoded Array:")
print(json.dumps(data, indent=2))
```

```python
# Decode nested object
toon_str = """user:
  name: Bob
  address:
    city: NYC
    zip: "10001"
    """

data = decoder.decode(toon_str)
print("Decoded Nested Object:")
print(json.dumps(data, indent=2))
```

## 5. Configuration Options <a id='config'></a>

TOON supports various configuration options for formatting.

```python
# Default configuration
config_default = ToonConfig()
encoder_default = ToonEncoder(config_default)

data = {"name": "test", "value": 123}
print("Default Config:")
print(encoder_default.encode(data))
```

```python
# Quoted strings configuration
config_quoted = ToonConfig(quote_strings=True)
encoder_quoted = ToonEncoder(config_quoted)

print("\nQuoted Strings Config:")
print(encoder_quoted.encode(data))
```

```python
# Custom indentation
config_indent4 = ToonConfig(indent_size=4)
encoder_indent4 = ToonEncoder(config_indent4)

nested_data = {"outer": {"inner": "value"}}
print("\n4-Space Indent Config:")
print(encoder_indent4.encode(nested_data))
```

```python
# Custom separator
config_sep = ToonConfig(separator= " = ")
encoder_sep = ToonEncoder(config_sep)

print("\nCustom Separator Config:")
print(encoder_sep.encode(data))
```

## 6. Edge Cases <a id='edge-cases'></a>

Testing various edge cases to ensure robustness.

```python
# Whitespace preservation
data = {
    "trailing": "hello ",
    "leading": " world",
    "both": " test "
}

toon = encoder.encode(data)
result = decoder.decode(toon)

print("Whitespace Preservation Test:")
print(f"Original: {data}")
print(f"\nTOON:\n{toon}")
print(f"\nDecoded: {result}")
print(f"\nMatch: {data == result} âœ“" if data == result else f"\nMatch: {data == result} âœ—")
```

## 10. Performance Testing <a id='performance'></a>

Comparing token usage between JSON and TOON formats.

```python
# Performance comparison
sample_data = {
    "users": [
        {"id": 1, "name": "Alice", "email": "alice@example.com", "active": True},
        {"id": 2, "name": "Bob", "email": "bob@example.com", "active": False},
        {"id": 3, "name": "Charlie", "email": "charlie@example.com", "active": True}
    ],
    "metadata": {
        "version": "1.0",
        "timestamp": "2025-12-09",
        "count": 3
    }
}

# Get stats
stats = get_conversion_stats(sample_data)

print("Performance Comparison:")
print(f"\nJSON Length: {stats['json_length']} characters")
print(f"TOON Length: {stats['toon_length']} characters")
print(f"\nJSON Tokens: {stats['json_tokens']}")
print(f"TOON Tokens: {stats['toon_tokens']}")
print(f"\nToken Reduction: {stats['token_reduction']:.1f}%")
print(f"\nâœ“ TOON is more compact!" if stats['token_reduction'] > 0 else "")
```

## 13. CLI Usage <a id='cli'></a>

Examples of using the command-line interface.

```python
import subprocess
import tempfile
import os

# Create temporary test file
test_data = {"name": "CLI Test", "value": 123, "items": [1, 2, 3]}

with tempfile.NamedTemporaryFile(mode='w', suffix='.json', delete=False, encoding='utf-8') as f:
    json.dump(test_data, f)
    temp_json = f.name

try:
    # Test to-toon command
    print("CLI Commands:")
    print("="*70)
    print("\n1. json2toon to-toon <file>")
    result = subprocess.run(
        ['json2toon', 'to-toon', temp_json],
        capture_output=True,
        text=True,
        encoding='utf-8'
    )
    if result.returncode == 0:
        print("âœ“ Command executed successfully")
        print(f"Output:\n{result.stdout}")
    else:
        print(f"âœ— Command failed: {result.stderr}")
        
finally:
    # Cleanup
    if os.path.exists(temp_json):
        os.unlink(temp_json)
```

```python
# Show CLI help
result = subprocess.run(
    ['json2toon', '--help'],
    capture_output=True,
    text=True,
    encoding='utf-8'
)

print("\n2. json2toon --help")
print(result.stdout)
```
